```python
import pandas as pd
import numpy as np

# CSV 불러오기
df = pd.read_csv("data/은마서_도로망/마포_서대문_은평_도로망_최종.csv")

# log_length
df["log_length"] = np.log1p(pd.to_numeric(df["length"], errors="coerce").clip(lower=0))

# abs_slope
df["abs_slope"] = pd.to_numeric(df["avg_slope"], errors="coerce").abs()

# len_x_slope
df["len_x_slope"] = pd.to_numeric(df["length"], errors="coerce").clip(lower=0) * \
                    pd.to_numeric(df["avg_slope"], errors="coerce").abs()

# 저장
df.to_csv("data/은마서_도로망/마포_서대문_은평_도로망_최종_v2.csv", index=False, encoding="utf-8-sig")

print("✅ 새로운 컬럼(log_length, abs_slope, len_x_slope) 추가 완료!")
```
